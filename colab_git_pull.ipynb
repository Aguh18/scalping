```python
# Step 6: Show results and download
print("\nüìà Step 6: Showing results...")
import joblib
import os
import zipfile

# Check if evaluation results exist
if os.path.exists('models/evaluation_results.pkl'):
    results = joblib.load('models/evaluation_results.pkl')
    print("üìä Model Performance:")
    print(f"Accuracy: {results['accuracy']:.4f}")
    print(f"Precision: {results['precision']:.4f}")
    print(f"Recall: {results['recall']:.4f}")
    print(f"F1-Score: {results['f1_score']:.4f}")
    
    # Show signal distribution
    if 'signal_distribution' in results:
        print("\nüìä Signal Distribution:")
        for signal, count in results['signal_distribution'].items():
            print(f"{signal}: {count}")
else:
    print("‚ùå Evaluation results not found.")
    print("\nüìÅ Available model files:")
    !ls -la models/

# Create results package
print("\nüì¶ Creating results package...")
with zipfile.ZipFile('bitcoin_scalping_results.zip', 'w') as zipf:
    # Add model files
    for root, dirs, files in os.walk('models'):
        for file in files:
            file_path = os.path.join(root, file)
            zipf.write(file_path, file)
    
    # Add source code
    for root, dirs, files in os.walk('src'):
        for file in files:
            file_path = os.path.join(root, file)
            zipf.write(file_path, file)
    
    # Add main files
    main_files = ['main.py', 'requirements.txt', 'README.md']
    for file in main_files:
        if os.path.exists(file):
            zipf.write(file, file)

print("‚úÖ Results package created: bitcoin_scalping_results.zip")

# Download the zip file
from google.colab import files
files.download('bitcoin_scalping_results.zip')

print("\nüéâ Git pull & run completed successfully!")
print("üìä Model is ready for trading!")
print("üì• Results downloaded to your computer!")
print("\n‚úÖ GIT PULL: Repository cloned from GitHub!")
print("‚úÖ DEPENDENCIES: Installed from requirements.txt!")
print("‚úÖ DATA: Already included in repository!")
print("‚úÖ TRAINING: Completed with GPU acceleration!")
print("‚úÖ READY: Model ready for trading!")
```
